```python
#extracting the year with parenthesis that are stored between parenthesis inside the 'title' and 
#store that in a variable 'year' 
movie['year']=movie.title.str.extract('(\(\d\d\d\d\))',expand=False)
movie.head(2)
```

```python
#replacing the 'year' with space or removing the 'year' from the 'title' column
movie['title']=movie.title.str.replace('(\(\d\d\d\d\))',' ')
movie.head(2)
```

```python
#Applying the strip function to get rid of any ending whitespace characters that may have appeared
movie['title'] = movie.title.str.strip()
```

```python
# removing the parenthesis from the 'year'
movie['year']=movie.year.str.extract('(\d\d\d\d)',expand=False)
movie.head(2)
```

```python
#removing the '|' from 'genres' using spilt function
movie['genres'] = movie.genres.str.split('|')
movie.head(2)
```

```python
data=rating.merge(movie,on='movieId')
data
```

```python
# groupby the 'title' column and count rest of column based on this groupby
data.groupby('title').count()
```

```python
# groupby the 'title' column and count the 'rating' column based on this groupby
movie_num_ratings=data.groupby('title')['rating'].count().reset_index()
movie_num_ratings
```

```python
movie_num_ratings.rename(columns={'rating':'num_ratings'},inplace=True)
```

```python
movie_num_ratings=movie_num_ratings[movie_num_ratings['num_ratings']>=50]
movie_num_ratings
```

```python
final_rating=data.merge(movie_num_ratings,on='title')
final_rating
```

```python
final_rating=final_rating.drop_duplicates(subset=['userId','title'])
final_rating
```

```python
movie_pivot=final_rating.pivot_table(columns='userId',index='title',values='rating')
movie_pivot
```

```python
movie_pivot=movie_pivot.fillna(0)
movie_pivot
```

```python
from scipy.sparse import csr_matrix
movie_matrix=csr_matrix(movie_pivot)
movie_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(movie_matrix)
```

```python
movie_pivot.index[237]
```

```python
distance,suggestion=model.kneighbors(movie_pivot.iloc[1,:].values.reshape(1,-1),n_neighbors=5)
```

```python
def recommend(moviename):
    m_id=np.where(movie_pivot.index==moviename)[0][0]
    distance,suggestion=model.kneighbors(movie_pivot.iloc[m_id,:].values.reshape(1,-1),n_neighbors=5)
    for i in suggestion[0]:
        print(movie_pivot.index[i])
```

```python
recommend('Twilight')
```
